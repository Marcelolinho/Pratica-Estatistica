# Porcentagem e Momentos #

### Porcentagem ###

- Consegue funilar uma quantidade em um valor de 0 a 100, sendo cada número uma quantia padrão
    - Exemplo: Uma cidade tem 500.000 habitantes, cada porcentagem equivale a 5.00
- Quartis
    - É quando dividimos uma certa quantidade de dados em quatro partes iguais, por exemplo 100 em 4 partes seria 25%

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0, 0.5, 10000)

plt.hist(data, 50)
plt.show()
```

- Funções que manipulam a porcentagem

```python
np.percentile(data, 50) # São passados os dados e o valor da porcentagem que deseja retornar
```

```python
np.percentile(data, 50, keepdims=True) # Caso a axis seja alterada, as dimensões ainda são as mesmas
```

### Momentos ###

- Fazendo os 4 momentos da estatística de momentos

```python
dados = np.random.normal(0, 0.5, 10000)

plt.hist(dados,50)
plt.show()
```

```python
np.mean(dados)
```

```python
np.var(dados)
```

```python
import scipy.stats as sp

sp.skew(dados)
```

```python
sp.kurtosis(dados)
```
