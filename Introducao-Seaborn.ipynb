# Introdução ao Seaborn (Matplotlib++) #

```python
import pandas as pd

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")

motores = df['# Gears'].value_counts()

motores.plot(kind='bar')
```

```python
import seaborn as sns
sns.set()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df)
```

```python
ax=sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

### Praticando ###

```python
sns.scatterplot(x='CombMPG', y='# Gears', data=df)
```

```python
sns.lmplot(x='CombMPG', y='# Gears', data=df)
```

```python
sns.boxplot(x='CombMPG', y='# Gears', data=df)
```
